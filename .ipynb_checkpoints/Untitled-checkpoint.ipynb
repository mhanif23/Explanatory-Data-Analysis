```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataApk = pd.read_csv("G:\FileNif\DataMining\diskominfo-od_17142_jml_layanan_aplikasi_umum_aplikasi_khusus__perangkat_d_data.csv")
dataApk.head()
```

```python
dataHost = pd.read_csv("G:\FileNif\DataMining\diskominfo-od_18163_jumlah_layanan_hosting_berdasarkan_perangkat_daerah_data.csv")
dataHost.head()
```

```python
dataApk.info()
dataHost.info()
```

```python
dataApk = dataApk.rename(columns = str.lower)
dataHost = dataHost.rename(columns = str.lower)
```

```python
dataApk.columns = dataApk.columns.str.replace(' ','_', regex=True)
dataHost.columns = dataHost.columns.str.replace(' ','_', regex=True)
```

```python
dataApk.isna().sum()
```

```python
dataHost.isna().sum()
```

```python
dataHost.rename(columns={'nama_perangkat_daerah': 'perangkat_daerah'}, inplace=True)
```

```python
dataMerge = pd.merge(dataApk, dataHost, how = "left", on=["perangkat_daerah"])
```

```python
dataMerge.info()
```

```python
deleteColumn = ["id_x", "kode_provinsi_y", "nama_provinsi_y", "satuan_y"]
dataMerge = dataMerge.drop(columns = deleteColumn)
```

```python
dataMerge = dataMerge.drop(columns = 'id_y')
```

```python
dataMerge.rename(columns={'kode_provinsi_x' : 'kode_provinsi', 'nama_provinsi_x' : 'nama_provinsi', 'satuan_x' : 'satuan', 'tahun_x' : 'tahun_aplikasi', 'tahun_y' : 'tahun_hosting'}, inplace=True)
```

```python
dataMerge.info()
```

```python
dataMerge['jumlah_layanan_hosting'] = dataMerge['jumlah_layanan_hosting'].fillna(0).astype(np.int64)
dataMerge['tahun_hosting'] = dataMerge['tahun_hosting'].fillna(0).astype(np.int64)
```

```python
dataMerge.isna().sum()
```

```python
dataMerge.head(30)
```

```python
dataMerge.boxplot(by ='tahun_aplikasi', column =['jumlah_aplikasi'], grid = False)
```

```python
dataMerge.boxplot(by ='tahun_hosting', column =['jumlah_layanan_hosting'], grid = False)
```

```python
dataMerge["gender"] = dataMerge["gender"].astype("category")
```

```python
dataMerge['tahun_hosting'] = dataMerge['tahun_hosting'].replace({0: 1})
```
